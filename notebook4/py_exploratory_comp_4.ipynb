As you may have noticed by now, we have created arrays with the correct shape and then filled them. Sometimes it is not easy (or possible) to determine the shape of an array beforehand. In that case we need to grow the array as we go. Changing the size of an array is not recommended in Python, although it is possible. A better way is to store values in a list and let the list grow. Then once the list is filled, the list can be converted into an array. For example, let's compute all perfect squares (integers that are the square of another integer) up to 200. We can start with an empty list and add a value to the list with the `append` command once we have found one:

```python
perfect_squares = []
for x in range(1,201):
    if int(sqrt(x)) * int(sqrt(x)) == x:
        perfect_squares.append(x)
perfect_squares = array(perfect_squares) # If you want to convert to array
print perfect_squares
```

```python
data = array([[1,2,3,5],
              [4,8,6,4],
              [3,5,4,6]])
print amax(data) # Finds the max value of the entire array
print amax(data,1) # Finds the max value along dimension 1
print argmax(data) # Finds the index of the max value of entire array
print argmax(data,1) # Finds the column index with the max value along dimension 1
print unravel_index( argmax(data),(3,4) ) # Gives i,j index of max for array of size (3,4)
```

###Keyword arguments
Functions may have multiple input arguments followed by keyword arguments. Arguments *must* be entered and must be entered in the order defined. Keyword arguments don't need to be entered. When they are not entered, the default value is used. Keyword arguments may be given in any order as long as they come after the regular arguments. If you specify the keyword arguments in the order they are defined in the argument list, you don't need to preceed them with the keyword, but it is saver and makes your code more readable. For example, the function you wrote for Exerise 3 can be written with a keyword argument for `alpha` as follows.

```python
def test(x,alpha=0.1):
    return exp(-alpha*x) * cos(x)
print test(3)  # Uses default alpha=0.1
print test(3,0.1)  # Same as default
print test(3,alpha=0.1)  # Same as default
print test(3,alpha=0.2)  # Now alpha=0.2
print test(3,0.2)  # Works as well, but the command above is more readable
```

###Local variables
Variables declared inside a function can only be used inside that function. The outside of a function doesn't know about the variables used inside the function. The only thing the outside knows is whatever the function returns. The only values from outside the function that the function can use are the values given to the function in its argument list. For example:

```python
def functest(x):
    a = 3
    b = 5
    return a*x + b
print functest(4)
a = 7
print functest(4) ## Will give the same value as before
# To change the value of a, add it to the argument list
def functest(x, a=3):
    b = 5
    return a*x + b
print functest(4,a=3)
print functest(4,a=7)
```

Note that the names used by the arguments of a function are the names used inside the function. They have no relation to the names used outside the function

```python
def functest(x, a=3):
    b = 5
    return a*x + b
y = 4
print functest(y)  # Inside function x will have the value of y
```

###Flow around a cylinder
Consider two-dimensional inviscid fluid flow (potential flow) around a cylinder.
The origin of the coordinate system is at the center of the cylinder.
The stream function is a function that is constant along stream lines. 
The stream function $\psi$ as a function of polar coordinates $r$ and $\theta$ is

$\psi = U(r-R^2/r)\sin(\theta)$

where $U$ is the flow in the $x$-direction, $r$ is the radial distance from the center of the cylinder, $\theta$ is the angle, and $R$ is the radius of the cylinder. Python has a very cool function to compute the angle given the $x$ and $y$ coordinate. The function is `arctan2(y,x)`. This function gives the correct angle (in radians, which is what you want) for any `y` and `x`. Note that the function takes as its *first* argument `y` and as its *second* argument `x`. The stream function is constant and equal to zero on the cylinder and doesn't really exist inside the cylinder, so let's make it zero there as well. 

***
###Exercise 4, Stream function for flow around a cylinder
Write a function that compute the stream function for flow around a cylinder. The function should take two arguments `x` and `y` and two keyword arguments, `U` and `R`, and should return the stream function value. If you write the function correctly, it should give psi(2, 4, U=1.0, R=1.0) = 3.8, and psi(0.5, 0, U=1.0, R=1.0) = 0 (inside the cylinder).

```python
def psi(x, y, U=1.0, R=1.0):
    r = sqrt( x**2 + y**2 )
    if r < R:
        rv = 0.0
    else:
        theta = arctan2(y,x)
        rv = U * (r - R**2 / r) * sin(theta)
    return rv

print psi(2, 4, U=1.0, R=1.0)
print psi(0.5, 0, U=1.0, R=1.0)
```

###Vectorization of a function
Not all functions can be called with an array of values as input argument. For example, the funciton `func` defined earlier in this notebook doesn't work with an array of `x` values. Trye it out:

```python
x = linspace(-6,6,100)
#y = func(x) # Run this line after removing the # to see the error that occurs. Then put the # back
```

The reason this doesn't work is that Python doesn't know what to do with the line `if x < 0` when `x` contains many values. Hence the error message `The truth value of an array with more than one element is ambiguous`. For some values of `x` the `if` statement may be `True`, for others it may be `False`. A simple way around this problem is to vectorize the function. That means we create a new function, let's call it `funcvec`, that is a vectorized form of `func` and can be called with an array as an argument (this is by far the easiest but not necessarily the computationally fastest way to make sure a function can be called with an array as an argument)

```python
funcvec = vectorize(func)
x = linspace(-6,6,100)
y = funcvec(x)
plot(x,y)
```

Take the first derivative of $f(x)$ and implement it in a Python function. Test your function by approximating the derivative with a finite difference formula and comparing this approximate value with the true value.

$\frac{\text{d}f}{\text{d}x}\approx \frac{f(x+d)-f(x-d)}{2d}$

where $d$ is a small number. Test your function for both $x<0$ and $x>0$.

```python
def dfuncdx(x):
    if x < 0:
        rv = -sin(x)
    else:
        rv = -exp(-x)
    return rv
d = 1e-6
x = -1
dfdx = (func(x+d) - func(x-d)) / (2*d)
print 'True value   ',dfuncdx(x)
print 'Approx value ',dfdx
x = 1
dfdx = (func(x+d) - func(x-d)) / (2*d)
print 'True value   ',dfuncdx(x)
print 'Approx value ',dfdx
```
